## Modelling the data set

### Splitting the Data set

```python
X = df.drop(columns='Class')
y = df['Class']
```

```python
under = NearMiss(n_neighbors=2)
```

```python
X_under, y_under = under.fit_resample(X, y, )
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_under, y_under, test_size = 0.10, random_state = 42)
```

```python
labels = 'Not Fraud', 'Fraud'
explode = (0, 0)

fig1, ax1 = plt.subplots()
ax1.pie(y_under.value_counts(), explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=60)
ax1.axis('equal')
plt.show()
```

```python
model = LogisticRegression(random_state = 42)
model.fit(X_train, y_train)
score = model.score(X_test, y_test)
score
```

## Evaluation

```python
y_preds = model.predict(X_test)
print(classification_report(y_test, y_preds))
```

```python
plot_confusion_matrix(model, X_test, y_test,
                     cmap=plt.cm.Blues);
```

```python
plot_roc_curve(model, X_test, y_test)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic Curve');
```

```python
RandomizedSearchCV.get_params(model)
```

```python
params = {
        'C': [0.1, 1.0, 5, 10],
         'max_iter': [20,50,100],
         'tol': [0.0001, 0.001, 0.01, 0.1, 1],
         }

search = RandomizedSearchCV(model, 
                            params,
                            n_iter=10,
                            scoring='accuracy',
                            n_jobs=-1,
                            cv=5,
                           )
```

```python
search.fit(X_train, y_train).set_params()
print(search.best_score_)
```
